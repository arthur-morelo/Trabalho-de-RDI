### Painel Interativo de Mercado

Abaixo está o cruzamento geográfico entre o volume de vagas ativas raspadas e a média salarial declarada pelos profissionais de dados. **Passe o mouse sobre os gráficos para explorar os detalhes.**

import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# Dados extraídos do Join Analítico
dados = {
    'UF': ['SP', 'PR', 'RJ', 'DF', 'MG'],
    'Volume_Vagas': [64, 6, 4, 2, 2],
    'Media_Salarial': [9448.41, 8406.50, 7171.59, 10656.71, 7382.81]
}
df_join = pd.DataFrame(dados).sort_values(by='Volume_Vagas', ascending=False)

# Criando layout lado a lado
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("Demanda de Mercado (Volume de Vagas)", "Compensação Financeira (Média Salarial)"),
    horizontal_spacing=0.15
)

# Gráfico 1: Vagas
fig.add_trace(
    go.Bar(
        x=df_join['UF'],
        y=df_join['Volume_Vagas'],
        name="Vagas",
        marker_color='#2c3e50',
        hovertemplate="<b>Estado:</b> %{x}<br><b>Vagas Coletadas:</b> %{y}<extra></extra>"
    ),
    row=1, col=1
)

# Gráfico 2: Salário
fig.add_trace(
    go.Scatter(
        x=df_join['UF'],
        y=df_join['Media_Salarial'],
        name="Média Salarial",
        mode="lines+markers",
        line=dict(color='#e74c3c', width=3),
        marker=dict(size=10),
        hovertemplate="<b>Estado:</b> %{x}<br><b>Média:</b> R$ %{y:,.2f}<extra></extra>"
    ),
    row=1, col=2
)

# Customização do Layout
fig.update_layout(
    template="plotly_white",
    height=500,
    showlegend=False,
    margin=dict(l=40, r=40, t=60, b=40)
)

fig.update_yaxes(title_text="Quantidade", row=1, col=1)
fig.update_yaxes(title_text="Reais (R$)", tickformat=",.0f", row=1, col=2)
